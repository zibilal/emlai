# Inisialisasi Figure dan Axes

```python
# X = np.linspace(0, 10, 10)
X = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sqrt(X))
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sin(X))
ax.plot(X, np.cos(X))
```

### Shorcut available for simple figure

```python
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X))
plt.plot(X, np.cos(X))
```

### Change color and style

```python
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X), 'r')
plt.plot(X, np.cos(X), 'b')
```

### Change line style

```python
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X), 'r--', linewidth=1)
plt.plot(X, np.cos(X), 'bo', markersize=3)
```

    'b'    # blue markers with default shape
    'ro'   # red circles
    'g-'   # green solid line
    '--'   # dashed line with default color
    'k^:'  # black triangle_up markers connected by a dotted line
    'c+-'  # cyan plus-sign markers connected by a solid line    

### Axis limit

```python
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X), 'r--', linewidth=1)
plt.plot(X, np.cos(X), 'bo', markersize=3)
plt.xlim(0, 15)
plt.ylim(-2, 2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X), 'r--', linewidth=1)
plt.plot(X, np.cos(X), 'bo', markersize=3)
plt.axis("equal")
```

# Back to non-shortcut way

```python
# copy dari 2 cell sebelumnya
plt.figure(figsize=(8,6))
plt.plot(X, np.sin(X), 'r--', linewidth=1)
plt.plot(X, np.cos(X), 'bo', markersize=3)
plt.xlim(0, 15)
plt.ylim(-2, 2)
plt.title("2 fungsi sinus")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sin(X), 'r--', linewidth=1)
ax.plot(X, np.cos(X), 'bo', markersize=3)
ax.xlim(0, 15)
ax.ylim(-2, 2)
ax.title("2 fungsi trig")
ax.xlabel("x")
ax.ylabel("y")
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sin(X), 'r--', linewidth=1)
ax.plot(X, np.cos(X), 'bo', markersize=3)
ax.set_xlim(0, 15)
ax.set_ylim(-2, 2)
ax.set_title("2 fungsi trig")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sin(X), 'r--', linewidth=1)
ax.plot(X, np.cos(X), 'bo', markersize=3)
ax.set(xlim=(0,15), ylim=(-2,2), xlabel="x", ylabel="y", title="2 fungsi trig")
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.plot(X, np.sin(X), 'r--', linewidth=1, label="sin(x)")
ax.plot(X, np.cos(X), 'bo', markersize=3, label="cos(x)")
ax.set(xlim=(0,15), ylim=(-2,2), xlabel="x", ylabel="y", title="2 fungsi trig")
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

# 3D plot

```python
# from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def f(x, y):
    return np.sqrt(x ** 2 + y ** 2)

x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, edgecolor='none', cmap="winter")
# ax.contour3D(X, Y, Z, 20)
# ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# ax.view_init(elev=60, azim=-45)
```

```python
df = pd.read_csv("cereal.csv", index_col="name")
df.head()
```

```python
sns.pairplot(df.dropna())
```

```python
df.shelf.value_counts()
```

```python
sns.pairplot(df.dropna(), hue="shelf", vars=["calories", "fat", "rating"])
# sns.pairplot(df.dropna(), hue="shelf", vars=["calories", "fat", "rating"], kind='reg')
```
