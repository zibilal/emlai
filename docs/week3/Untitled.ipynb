```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# object oriented method
# create figure object
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('The Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes1.set_title('LARGER Plots')
axes2.plot(y, x)
axes2.set_title('SMALLER Plots')
```

```python
fig = plt.figure()
testax = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('First Plot')
axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
ax[0].plot(x, y)
ax[1].plot(y, x)
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X squared')
ax.plot(x, x**3, label='X cubed')

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=3, ls='-', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```
