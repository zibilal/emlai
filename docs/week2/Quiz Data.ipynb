## Soal 2
- load dataset: `weight-height.csv`
- inspect
- plot menggunakan scatter plot dengan Weight sebagai funsi dari Height
- plot populasi male dan female menggunakan 2 warna yang berbeda pada scatter plot baru
- kasih label juga axisnya!

```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/weight-height.csv')
```

```python
df
```

```python
df.plot(kind='scatter', x='Weight', y='Height')
```

```python
males = df[df['Gender'] == 'Male']
females = df[df['Gender'] == 'Female']

fig, ax = plt.subplots()
males.plot(kind='scatter', x='Weight', y='Height', alpha=0.6, color='blue', ax=ax, title='Male')
females.plot(kind='scatter', x='Weight', y='Height', alpha=0.6, color='red', ax=ax, title='Female')
```

## Soal 3
- plot histogram dari tinggi males dan females untuk plot yang sama 
- gunakan alpha untuk mngontrol transparansi dari plot
- plot garis vertikal dari rata-rata untuk tiap populasi menggunakan `plt.axvline()` 

```python
fig, ax = plt.subplots()
males.plot(kind='hist',bins=50, x='Weight', y='Height', alpha=0.6, color='blue', ax=ax)
females.plot(kind='hist',bins=50, x='Weight', y='Height', alpha=0.6, color='red', title='Gender', ax=ax)
plt.axvline(males['Height'].mean(), color='blue')
plt.axvline(females['Height'].mean(), color='red')
```
