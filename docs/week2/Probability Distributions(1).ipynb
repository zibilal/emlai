```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
%matplotlib inline
```

```python
mean = 100
stdev = 15
lower = 120
upper = np.inf
```

```python
x = np.linspace(-4, 4, 100) * stdev + mean
prob = norm.pdf(x, mean, stdev)
```

```python
fig = plt.figure(figsize=(10,6))
fig.suptitle('Normal Distribution of IQ Scores',fontsize=14,fontweight='bold')

ax = fig.add_subplot(111)
ax.plot(x,prob,'k')
ax.set_xlabel('IQ Values')
ax.set_xlim((x.min(),x.max()))
ax.set_ylim((prob.min(),prob.max()))

ax.get_yaxis().set_visible(False)
for key in ax.spines.keys():
    if key in {'top','left','right'}:
        ax.spines[key].set_visible(False)

i = (x >= lower) & (x <= upper)
verts = [(lower,0),*zip(x[i],prob[i]),(upper,0)]
poly = Polygon(verts, facecolor='r', edgecolor='b')
ax.add_patch(poly)

area = 1 - norm.cdf(lower, mean, stdev) 
ax.set_title("P({} < IQ < {}) = {:.3f}".format(lower,upper,area));
```
