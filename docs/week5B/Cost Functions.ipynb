```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('headbrain.csv')
df = df[['Head Size(cm^3)','Brain Weight(grams)']]
df.head()
```

```python
df.shape
```

```python
df.plot(kind='scatter',
        x='Head Size(cm^3)',
        y='Brain Weight(grams)',
        title='Head Size vs Brain Weights')

# Here we're plotting the red line 'by hand' with fixed values
# We'll try to learn this line with an algorithm below
plt.plot([2750, 4750], [1000, 1700], color='red', linewidth=3)
```

```python
def line(x, w=0, b=0):
    return x * w + b
```

```python
x = np.linspace(2750, 4750, 200)
```

```python
x
```

```python
yhat = line(x, w=0, b=0)
```

```python
df.plot(kind='scatter',
        x='Head Size(cm^3)',
        y='Brain Weight(grams)',
        title='Head Size vs Brain Weights')
plt.plot(x, yhat, color='red', linewidth=3)
```

__Cost Function__

```python
def mean_squared_error(y_true, y_pred):
    s = (y_true - y_pred)**2
    return s.mean()
```

```python
X = df[['Head Size(cm^3)']].values
y_true = df['Brain Weight(grams)'].values
```

```python
y_true
```

```python
y_pred = line(X)
```

```python
y_pred
```

```python
mean_squared_error(y_true, y_pred.ravel())
```

```python
plt.figure(figsize=(10,5))

# we are going to draw 2 plots in the same figure
# first plot, data and a few lines
ax1 = plt.subplot(121)
df.plot(kind='scatter',
        x='Head Size(cm^3)',
        y='Brain Weight(grams)',
        title='Head Size vs Brain Weights',ax=ax1)

# let's explore the cost function for a few values of b betweeb -100 and +150 
bbs = np.array([-2500,-2000, -1000, 0, 1000, 1500])
mses = []
for b in bbs:
    y_pred = line(X, w=0.5, b=b)
    mse = mean_squared_error(y_true, y_pred)
    mses.append(mse)
    plt.plot(X, y_pred)

# second plot: Cost function
ax2 = plt.subplot(122)
plt.plot(bbs, mses, 'o-')
plt.title('Cost as function of b')
plt.xlabel('b')
```
